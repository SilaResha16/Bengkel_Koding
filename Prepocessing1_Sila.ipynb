# **Preprocessing**

**Remove missing values**

```python
df = df.dropna()
```

```python
df.info()
```

**Delete duplicate data**

```python
df = df.drop_duplicates()
```

```python
df.info()
```

**Check unique values**

```python
for col in df.columns:
    print(f"Kolom: {col} - Jumlah Nilai Unik: {df[col].nunique()}")
    print(f"Nilai unik: {df[col].unique()}")
    print("-" * 40)
```

```python
kolom_unik = ["Gender", "CALC", "FAVC", "SCC", "SMOKE", "family_history_with_overweight", "CAEC", "MTRANS", "NObeyesdad"]

for col in kolom_unik:
    print(f"Kolom: {col} - Jumlah Nilai Unik: {df[col].nunique()}")
    print(f"Nilai unik: {df[col].unique()}")
    print("-" * 40)
```

```python
df= df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

df.replace("?", np.nan, inplace=True)

for col in df.columns:
    if df[col].dtype == "object":
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)

print("Jumlah data yang masih memiliki NaN setelah imputasi:\n", df.isnull().sum())
```

**Handle outlier**

```python
df.describe()
```

```python
for col in numerical_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]
```
